```python
import pandas as pd
import numpy as np
import statistics as stats
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
customer_data = pd.read_csv("Data/round_2_cleaned.csv")
print(customer_data.info())
```

## Activity 3
- Which columns are numerical?
- Which columns are categorical?
- Datetime format : Extract the months from the dataset and store in a separate column. Then filter the data to show only the information for the first quarter , ie. January, February and March. Hint: If data from March does not exist, consider only January and February.

```python
#Which columns are numerical?
print(list(customer_data.select_dtypes(include='number').columns.values))

#Which columns are categorical?
print(list(customer_data.select_dtypes(include='object').columns.values))
```

```python
#Datetime format
pd.to_datetime(customer_data["effective to date"],format="%m/%d/%y")
customer_data['month'] = pd.DatetimeIndex(customer_data['effective to date']).month
print(customer_data['month'])
customer_data.month.value_counts()
```

```python
#Then filter the data to show only the information for the first quarter , ie. January, February and March. 
#Hint: If data from March does not exist, consider only January and February.

customer_data = customer_data[customer_data['month'] <= 2]
print(customer_data)
```

```python
# I create the dataframes and I drop the columns that are not needed ( the column month should not be in the numerical df)

numerical = customer_data.select_dtypes(include='number')
numerical.drop(["month"],axis=1, inplace=True)
categorical = customer_data.select_dtypes(include='object')

print(numerical.info())
print(categorical.columns)
```

```python
#Now we will try to check the normality of the numerical variables visually
#Use seaborn library to construct distribution plots for the numerical variables
#Use Matplotlib to construct histograms
#Do the distributions for different numerical variables look like a normal distribution
```

```python
sns.set_style('whitegrid')
sns.displot(numerical['income'], kde = False, color ='red', bins = 30)
```

```python
# create the figure and axes
fig,ax = plt.subplots(2,4,figsize = (20,6)) # Here the figsize controls the full canvas size, not the individual figure size

# add an overall title for the figure
fig.suptitle('Customer Data Case Study with Seaborn', fontsize=18)

# create first plot (left), adding a title
sns.histplot(numerical['customer lifetime value'], kde = False, color ='red', bins = 30,ax =ax[0,0])
sns.histplot(numerical['income'], kde = False, color ='blue', bins = 30,ax =ax[0,1])
sns.histplot(numerical['monthly premium auto'], kde = False, color ='green', bins = 30,ax =ax[0,2])
sns.histplot(numerical['months since last claim'], kde = False, color ='orange', bins = 30,ax =ax[0,3])
sns.histplot(numerical['months since policy inception'], kde = False, color ='yellow', bins = 30,ax =ax[1,0])
sns.histplot(numerical['number of open complaints'], kde = False, color ='purple', bins = 30,ax =ax[1,1])
sns.histplot(numerical['number of policies'], kde = False, color ='pink', bins = 30,ax =ax[1,2])
sns.histplot(numerical['total claim amount'], kde = False, color ='brown', bins = 30,ax =ax[1,3])

```

```python
# create the figure and axes
fig2,axes2 = plt.subplots(2,4,figsize = (20,6)) # Here the figsize controls the full canvas size, not the individual figure size

# add an overall title for the figure
fig2.suptitle('Customer Data Study with Matplotlib', fontsize=18)

# create first plot (left), adding a title
axes2[0,0].hist(numerical['customer lifetime value'], color=('red'))
axes2[0,0].set_title('customer lifetime value', fontsize=15)
axes2[0,1].hist(numerical['income'], color=('blue'))
axes2[0,1].set_title('income', fontsize=15)
axes2[0,2].hist(numerical['monthly premium auto'], color=('green'))
axes2[0,2].set_title('monthly premium auto', fontsize=15)
axes2[0,3].hist(numerical['months since last claim'], color=('orange'))
axes2[0,3].set_title('months since last claim', fontsize=15)
axes2[1,0].hist(numerical['months since policy inception'], color=('yellow'))
axes2[1,0].set_title('months since policy inception', fontsize=15)
axes2[1,1].hist(numerical['number of open complaints'], color=('purple'))
axes2[1,1].set_title('number of open complaints', fontsize=15)
axes2[1,2].hist(numerical['number of policies'], color=('pink'))
axes2[1,2].set_title('number of policies', fontsize=15)
axes2[1,3].hist(numerical['total claim amount'], color=('brown'))
axes2[1,3].set_title('total claim amount', fontsize=15)

```

```python
#Show a plot of the response rate by the total claim amount

boxplot = customer_data.boxplot(column="total claim amount",by="response",figsize=(10,10))

customer_data["total claim amount"].hist(by=customer_data["response"], bins=[0,500,1000,1500,2000],figsize=(10,10))
```

```python
#Show a plot of the response rate by income

customer_data["income"].hist(by=customer_data["response"], bins=20,figsize=(10,10))
print((customer_data["income"] == 0).sum())
```

```python
corr = customer_data.corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(9, 7))
    ax = sns.heatmap(corr, mask=mask,cmap='coolwarm', vmin=-1,vmax=1,annot=True, square=True)
```
